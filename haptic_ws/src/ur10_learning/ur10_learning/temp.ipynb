```python
from math import pi, sin, cos

# Euler angles
roll = 0
pitch = 0
yaw = pi/2

# Conversion to quaternion
qx = sin(roll / 2) * cos(pitch / 2) * cos(yaw / 2) - cos(roll / 2) * sin(pitch / 2) * sin(yaw / 2)
qy = cos(roll / 2) * sin(pitch / 2) * cos(yaw / 2) + sin(roll / 2) * cos(pitch / 2) * sin(yaw / 2)
qz = cos(roll / 2) * cos(pitch / 2) * sin(yaw / 2) - sin(roll / 2) * sin(pitch / 2) * cos(yaw / 2)
qw = cos(roll / 2) * cos(pitch / 2) * cos(yaw / 2) + sin(roll / 2) * sin(pitch / 2) * sin(yaw / 2)

qx, qy, qz, qw
```

```python
import pickle
import os
import numpy as np
```

```python
folderpath = '~/ros_projects/haptic_ws/src/ur10_learning/data/demonstrations/'
filename = folderpath + 'env1/end_effector/demo4.pkl'
filename = os.path.expanduser(filename)

with open(filename, 'rb') as file:
    # Load the list of lists from the file
    end_effector_data = np.array(pickle.load(file))

x = np.array(end_effector_data)[:,0]
y = np.array(end_effector_data)[:,1]
```

```python
folderpath = '~/ros_projects/haptic_ws/src/ur10_learning/data/demonstrations/'
x_model = []
y_model = []
for i in range(5):
    filename = folderpath + 'env1/DAgger_model/trajectories/model'+str(i+1) + '_traj.pkl'
    filename = os.path.expanduser(filename)
    with open(filename, 'rb') as file:
        # Load the list of lists from the file
        end_effector_model_data = np.array(pickle.load(file))
    x_model.append(np.array(end_effector_model_data)[:,0])
    y_model.append(np.array(end_effector_model_data)[:,1])
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# Create a figure and an axes.
fig, ax = plt.subplots(figsize=(10, 6))

# Create a circle patch.
ball = Circle((0.3, 1.0), 0.025, edgecolor='#e56b6f', facecolor='#e56b6f', linewidth=2, label='Ball Position')
coke = Circle((-0.3, 1.0), 0.03, edgecolor='#b56576', facecolor='#b56576', linewidth=2, label='Coke Position')
bowl = Circle((-0.3, 0.5), 0.04, edgecolor='#06d6a0', facecolor='#06d6a0', linewidth=2, label='Bowl Position')

ax.plot(x, y, 'r--', color='#98c1d9', linewidth=2, label='Expert Demo')

i = 2
ax.plot(x_model[i], y_model[i], linewidth=2, label='Model Trajectory')  # Assuming x_model and y_model are defined.

# Add the circle to the axes
ax.add_patch(ball)
ax.add_patch(coke)
ax.add_patch(bowl)

# Add points
px = [0.3,-0.3,-0.305]
py = [1.0,1.0,0.5]
ax.plot(px,py, 'o', color='#073b4c', label='Critical points')


# Set limits and aspect of the axes
ax.set_xlim(-0.4, 0.4)
ax.set_ylim(0.35, 1.15)
ax.set_aspect('equal')

# Set axis labels
ax.set_xlabel('X Positions') 
ax.set_ylabel('Y Positions') 

ax.legend()



# Show the plot
plt.show()
```
